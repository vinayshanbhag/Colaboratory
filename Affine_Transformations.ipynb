```
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
import string

# points
a, b = (0, 0), (1, 0) 

# matrix with row vectors of points
M = np.array([a, b])

# Identity transformation matrix
I = np.eye(2)
```

```
def draw(vectors, names = string.ascii_uppercase):
  """draw vectors"""
  colors = 'rgbcmyk'#[color for color in mcolors.TABLEAU_COLORS]
  pts_x = []
  pts_y = []
  for idx, row in enumerate(vectors):
      x, y = row
      pts_x.append(x)
      pts_y.append(y)
      plt.scatter(x, y, color=colors[idx%len(colors)])
      plt.text(x + 0.15, y, f"{names[idx]} ({x:0.0f},{y:0.0f})")
  pts_x.append(pts_x[0])
  pts_y.append(pts_y[0])
  plt.plot(pts_x, pts_y, color="black", linestyle='dotted');
  plt.grid();
  plt.show();
```

```
draw(M @ I)
```

```
# Rotation matrix - rotate by theta radians
theta = np.pi/6 # 30 deg
R = np.array([[np.cos(theta), np.sin(theta)], [-np.sin(theta), np.cos(theta)]])
draw(M @ R, names=[f"{c}\'" for c in string.ascii_uppercase])
```

```
a,b,c,d = (1,1),(1,-1),(-1,-1),(-1,1)
M = [a,b,c,d]
```

```
draw(M)
```

```
# Rotation matrix - rotate by theta radians
theta = np.pi/3 # 60 deg
R = np.array([[np.cos(theta), np.sin(theta)], [-np.sin(theta), np.cos(theta)]])
draw(M @ R,names=[f"{c}\'" for c in string.ascii_uppercase])
```

```
# Reflection
Reflection = np.array([[-1,0],[0,1]])
draw(M @ Reflection,names=[f"{c}\'" for c in string.ascii_uppercase])
```

```
# Shear
Shear = np.array([[1,0.5],[0,1]])
draw(M @ Shear, names=[f"{c}\'" for c in string.ascii_uppercase])
```

```
Z = Shear @ R
draw(M @ Z, names=[f"{c}\'" for c in string.ascii_uppercase])
```

```
Z = R @ Shear
draw(M @ Z, names=[f"{c}\'" for c in string.ascii_uppercase])
```
