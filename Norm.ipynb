```
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import numpy as np
from numpy import sin, cos, tan, pi,inf, random, around
from numpy.linalg import norm
import string
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```
plt.figure(figsize=(8,8));
plt.gca().set_aspect('equal', adjustable='box');
for p in [1/2,1,2,inf]:
  pts = np.array([k for k in random.uniform(-1,1,(1000000,2)) if around(norm(k,p),2)==1])
  plt.text(max(pts[:,0])/3, np.median(abs(pts[:,1])*1.1), f"L({p}) norm")
  plt.scatter(pts[:,0],pts[:,1], marker=".");
```
