### Interpolation

Interpolation is a method of constructing new data points within the range of a discrete set of known data points.

```
import numpy as np
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
```

Consider these known data points

```
X=np.linspace(0, 10, num=11, endpoint=True)
Y=np.cos(-X**3)
plt.figure(figsize=(15,5))
plt.title("Known data points")
plt.plot(X,Y,'o',color='m');
```

Linear interpolation joins each consecutive data point by a straight line

```
X2 = np.linspace(0,10,100)
plt.figure(figsize=(15,5))
plt.title("Linear Interpolation")
plt.plot(X,Y,'o',color='m')
intrp = interp1d(X,Y, kind="linear")
plt.plot(X2,intrp(X2),'--')
plt.legend(['known data points','linear interpolation']);
```

```
X2 = np.linspace(0,10,100)
plt.figure(figsize=(15,5))
plt.title("Quadratic Interpolation")
plt.plot(X,Y,'o',color='m')
intrp = interp1d(X,Y, kind="quadratic")
plt.plot(X2,intrp(X2),'--')
plt.legend(['known data points','quadratic interpolation']);
```

```
kind = ['linear','quadratic','cubic']
X2 = np.linspace(0,10,100)
colors = ['r','g','b']
plt.figure(figsize=(15,5))
plt.title("Interpolation")
plt.plot(X,Y,'o',color='m')
for idx,k in enumerate(kind):
  intrp = interp1d(X,Y, kind=k)
  plt.plot(X2,intrp(X2),'--',color=colors[idx])

plt.legend(['known data points']+kind);
```

```
kind = ['zero','nearest','next']
X2 = np.linspace(0,10,100)
colors = ['r','g','b','cyan']
plt.figure(figsize=(15,5))
plt.title("Interpolation")
plt.plot(X,Y,'o',color='m')
for idx,k in enumerate(kind):
  intrp = interp1d(X,Y, kind=k)
  plt.plot(X2,intrp(X2),'--',color=colors[idx])

plt.legend(['known data points']+kind);
```
