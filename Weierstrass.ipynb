### Weierstrass function

The Weierstrass function is an example of a pathological real-valued function on the real line. The function has the property of being continuous everywhere but differentiable nowhere. Until Weierstrass published this in 1872, most mathematians believed that a continuous function could only fail to
be differentiable at some collection of isolated points. 

The function is defined as an infinite fourier series of cosine functions

$f(x) = \frac{\cos(3x\pi)}{2^1} + \frac{\cos(3^2x\pi)}{2^2} + \frac{\cos(3^3x\pi)}{2^3}\dots $

Approximating this in numpy...


```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(-1,1,1000)
y = np.zeros(1000)
for i in range(0,50):
  y = y + np.cos(3**i*np.pi*x)/2**i
plt.title("Weierstrass function");
plt.plot(x,y);
```
