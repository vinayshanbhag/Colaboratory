```
import matplotlib.pyplot as plt
%matplotlib inline
```

```
def primes(n):
    # Source: https://stackoverflow.com/questions/2068372/fastest-way-to-list-all-primes-below-n/3035188#3035188
    """ Returns  a list of primes < n """
    sieve = [True] * n
    for i in range(3,int(n**0.5)+1,2):
        if sieve[i]:
            sieve[i*i::2*i]=[False]*((n-i*i-1)//(2*i)+1)
    return [2] + [i for i in range(3,n,2) if sieve[i]]
```

```
n = 1_000_000
p = primes(n)
fig = plt.figure(figsize=(10,10))
fig.suptitle("a million primes", fontsize=20)
plt.polar(p,p,'.', markersize=0.5,color='g');
```
